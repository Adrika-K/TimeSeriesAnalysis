```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
DF_H = df["date"],df["humidity"]
DF_H = pd.DataFrame(DF_H).T
DF_H.head()
```

```python
DF_H.set_index("date" , inplace = True)
```

```python
DF_H.head()
```

```python
DF_H.index = pd.to_datetime(DF_H.index , dayfirst= True)
```

```python
plt.plot(DF_H)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
# Decomposing Time series
decomposition = seasonal_decompose(DF_H.humidity)  
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(50, 8)
```
